```python
import pandas as pd
```

```python
df_20 = pd.read_csv('/home/anuarsantoyo/PycharmProjects/Batbot/archive/experiments/optimizer_batbotV2_2D_BSQJNP8/230915/test_extended1/results.csv')
df_20['Generation'] = df_20['Generation'].astype(int)
df_20 = df_20.rename(columns={'Motor': 'Motor Power', 'Attack': 'Attack Angle'})

df_25 = pd.read_csv('/home/anuarsantoyo/PycharmProjects/Batbot/archive/experiments/optimizer_batbotV2_2D_BSQJNP8/230916/test_folding1/results.csv')
df_25['Generation'] = df_25['Generation'].astype(int)
df_25 = df_25.rename(columns={'Motor': 'Motor Power', 'Attack': 'Attack Angle'})

df_30 = pd.read_csv('/home/anuarsantoyo/PycharmProjects/Batbot/archive/experiments/optimizer_batbotV2_2D_BSQJNP8/230916/test_folding2/results.csv')
df_30['Generation'] = df_30['Generation'].astype(int)
df_30 = df_30.rename(columns={'Motor': 'Motor Power', 'Attack': 'Attack Angle'})
```

```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots

# Assuming 'df' is your DataFrame and is already loaded with the necessary data
df = df_30.copy()
# Create a figure with two subplots (1 row, 2 columns)
fig = make_subplots(rows=1, cols=2, subplot_titles=('Motor Power vs Score', 'Attack Angle vs Score'))

# Scatter plot for Motor Power vs Score
fig.add_trace(
    go.Scatter(x=df['Motor Power'], y=df['Score'], mode='markers', marker=dict(color=df['Generation'], showscale=True), text=df['Generation']),
    row=1, col=1
)

# Scatter plot for Attack Angle vs Score
fig.add_trace(
    go.Scatter(x=df['Attack Angle'], y=df['Score'], mode='markers', marker=dict(color=df['Generation'], showscale=True), text=df['Generation']),
    row=1, col=2
)

# Update x-axis and y-axis labels
fig.update_xaxes(title_text="Motor Power", row=1, col=1)
fig.update_xaxes(title_text="Attack Angle", row=1, col=2)
fig.update_yaxes(title_text="Score", row=1, col=1)
fig.update_yaxes(title_text="Score", row=1, col=2)

# Update layout (if needed) and show the figure
fig.update_layout(height=600, width=1000, title_text="Optimal attack angle and motor power using 30 generation of CMA")
fig.show()
```
