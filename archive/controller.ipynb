# Control Batbot with Xbox controller

```python
import plotly.express as px
import pandas as pd
from inputs import get_gamepad
import time
import numpy as np
import serial
```

```python
measurements = {'ABS_X':0,'ABS_Y':0,'ABS_Z':0}
pd_list = []
for i in range(10000):
    time.sleep(0.01)
    events = get_gamepad()
    for event in events:
        if event.ev_type == 'Absolute':
            if event.code in measurements:
                measurements[event.code] = event.state
                pd_list.append(measurements.copy())
                print(measurements)
```

```python
df = pd.DataFrame(pd_list)
np.interp(0.90, [0, 1], [260, 275])
```

```python
px.scatter_3d(df, x='ABS_X', y='ABS_Y', z='ABS_Z' )
```

```python
port = "/dev/ttyACM1"
ser = serial.Serial(port=port, baudrate=115200, bytesize=8, parity='N', stopbits=1)
measurements = {'ABS_X':0,'ABS_Y':0,'ABS_Z':0}
for i in range(10000):
    time.sleep(0.01)
    events = get_gamepad()  # Get the list of events
    for event in events:
        if event.ev_type == 'Absolute':
            if event.code in measurements:
                measurements[event.code] = event.state
                x = np.interp(measurements['ABS_X'],[-32768, 32767], [40,120])
                y = np.interp(measurements['ABS_Y'],[-32768, 32767], [150,30])              
                z = np.interp(measurements['ABS_Z'],[0, 255], [250,270])
                print(f'{x},{y},{z}')
                ser = serial.Serial(port=port, baudrate=115200, bytesize=8, parity='N', stopbits=1)
                ser.write(str.encode(f'{x},{y},{z}'))
```
