```python
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
# import required module
import os
# assign directory
directory = 'data/5deg'

experiments = {}
# that directory
for filename in os.listdir(directory):
    f = os.path.join(directory, filename)
    experiments[filename] = pd.read_csv(f)
    experiments[filename]['experiment'] = filename
```

```python
for exp in experiments:
    plt.plot(experiments[exp].timestamp,experiments[exp].lift, label=exp)
plt.legend()
plt.xlim(6,10)
plt.show()
```

```python
df = pd.concat([experiments[exp] for exp in experiments], ignore_index=True)
```

```python
fig = px.line(df, x='timestamp', y='lift', color='experiment')
fig.write_html("data/folding_influence_5deg.html")
```
