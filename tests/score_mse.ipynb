# Visualize

```python
df.plot(x='timestamp')
```

```python
for sensor in sensors_col:
    y = df[sensor]
    
    peaks, _ = find_peaks(y, height=0)
    valleys, _ = find_peaks(-y, height=0)
    plt.title(sensor)
    plt.plot(y)
    plt.plot(peaks, y[peaks], "x")
    plt.plot(valleys, y[valleys], "o")
    plt.show()
    
    print(len(peaks))
```
