```python
import numpy as np
```

```python
import random
import matplotlib.pyplot as plt
```

```python
total_of_points = 5000
in_circle = 0

for i in range(total_of_points):
    point = np.array([random.random(), random.random()])
    dif = np.array([0.5,0.5])- point
    distance = np.sqrt(dif[0]**2 + dif[1]**2)
    if distance<0.5:
        in_circle += 1
        plt.scatter(*point,c='r')
    else:
        plt.scatter(*point,c='b')
plt.show()
in_circle/total_of_points
```
